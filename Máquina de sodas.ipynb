#Maquina expendedora de sodas, dulces y frituras


```python
class Bebida:
  def __init__(self, nombre, cantidad, precio):
    self.nombre = nombre
    self.cantidad = cantidad
    self.precio = precio

class Dulce:
  def __init__(self, nombre, cantidad, precio):
    self.nombre = nombre
    self.cantidad = cantidad
    self.precio = precio

class Fritura:
  def __init__(self, nombre, cantidad, precio):
    self.nombre  = nombre
    self.cantidad = cantidad
    self.precio = precio

def comprar_bebida(bebida):
    saldo = 0
    print(f'Has elegido {bebida.nombre}, de {bebida.cantidad}')
    while saldo < bebida.precio:
        print(f'Te faltan: ${bebida.precio - saldo} pesos para tu bebida')
        dinero = int(input('Ingresa una moneda, solo acepto monedas de $1, $2, $5 y $10 pesos y billetes de $20, $50 y $100 pesos: '))
        if dinero in [1, 2, 5, 10, 20, 50, 100]:
            saldo += dinero
        else:
            print('Moneda o billete inválido. Introduce una moneda o billete válido.')
    print('Debes $0 pesos')
    print(f'Aquí tienes tu bebida, tu cambio es: ${saldo - bebida.precio} pesos. ¡Gracias, vuelva pronto!')

def comprar_dulce(dulce):
    saldo = 0
    print(f'Has elegido {dulce.nombre}, de {dulce.cantidad}')
    while saldo < dulce.precio:
        print(f'Te faltan: ${dulce.precio - saldo} pesos para tu dulce')
        dinero = int(input('Ingresa una moneda, solo acepto monedas de $1, $2, $5 y $10 pesos y billetes de $20, $50 y $100 pesos: '))
        if dinero in [1, 2, 5, 10, 20, 50, 100]:
            saldo += dinero
        else:
            print('Moneda o billete inválido. Introduce una moneda o billete válido.')
    print('Debes $0 pesos')
    print(f'Aquí tienes tu dulce, tu cambio es: ${saldo - dulce.precio} pesos. ¡Gracias, vuelva pronto!')

def comprar_fritura(fritura):
    saldo = 0
    print(f'Has elegido {fritura.nombre}, de {fritura.cantidad}')
    while saldo < fritura.precio:
        print(f'Te faltan: ${fritura.precio - saldo} pesos para tus frituras')
        dinero = int(input('Ingresa una moneda, solo acepto monedas de $1, $2, $5 y $10 pesos y billetes de $20, $50 y $100 pesos: '))
        if dinero in [1, 2, 5, 10, 20, 50, 100]:
            saldo += dinero
        else:
            print('Moneda o billete inválido. Introduce una moneda o billete válido.')
    print('Debes $0 pesos')
    print(f'Aquí tienes tus frituras, tu cambio es: ${saldo - fritura.precio} pesos. ¡Gracias, vuelva pronto!')


bebidas = [
    Bebida('Coca Cola', '600ml', 18),
    Bebida('Sprite', '600ml', 18),
    Bebida('Fanta', '600ml', 18),
    Bebida('Jugo Jumex Durazno', '600ml', 20),
    Bebida('Jugo Jumex Naranja', '600ml', 20),
    Bebida('Agua Ciel', '1Lt', 14),
    Bebida('Sidral Mundet', '600ml', 20),
    Bebida('Fresca', '600ml', 20),
    Bebida('Pepsi', '600ml', 18),
    Bebida('7up', '600ml', 18)
]


dulces = [
    Dulce('Snickers', '1 barra', 20),
    Dulce('Pelon pelorrico', '30 mg', 5),
    Dulce('Dulces pelon pelorrico', '30mg', 10),
    Dulce('Milkyway', '1 barra', 20),
    Dulce('Halls menta', '15 mg', 12),
    Dulce('Halls negras', '15 mg', 12),
    Dulce('Paleta bubbaloo', '1 pieza', 10),
    Dulce('Holly ranchers', '15 mg', 10),
    Dulce('Mentos', '15 mg', 12),
    Dulce('Hershey blanco', '1 barra', 20),
]

frituras = [
    Fritura('Sabritas sal', '65 mg', 15),
    Fritura('Sabritas adobadas', '65 mg', 15),
    Fritura('Doritos nacho', '65 mg', 15),
    Fritura('Takis verdes', '50 mg', 13),
    Fritura('Takis fuego', '50 mg', 15),
    Fritura('Ruffles', '65 mg', 15),
    Fritura('Rancheritos', '65 mg', 15),
    Fritura('Cheetos torciditos', '50 mg', 15),
    Fritura('Runners', '35 mg', 15),
    Fritura('Chipotles', '50 mg', 15),
]

while True:
  print('Bienvenido a la maquina Maxi Sodas, Dulces y Frituras')
  opcion = int(input("""Que opción deseas elegir:
              1: Bebidas
              2: Dulces
              3: Frituras
              4: Salir
                          """))
  if opcion == 1:

    while True:
        print("¿Qué deseas beber?")
        for i, bebida in enumerate(bebidas, start=1):
            print(f"{i}: {bebida.nombre} {bebida.cantidad} ${bebida.precio}")
        print(f"{len(bebidas) + 1}: Volver al menú principal")

        opcion = int(input("Seleccione una opción: "))

        if opcion == len(bebidas) + 1:
            print("¡Gracias vuelve pronto!")
            break

        if opcion < 1 or opcion > len(bebidas):
            print("Opción inválida, por favor ingrese una opción válida")
            continue

        bebida_seleccionada = bebidas[opcion - 1]
        comprar_bebida(bebida_seleccionada)

  elif opcion == 2:
    while True:
        print("¿Qué dulce te gustaría?")
        for i, dulce in enumerate(dulces, start=1):
            print(f"{i}: {dulce.nombre} {dulce.cantidad} ${dulce.precio}")
        print(f"{len(dulces) + 1}: Volver al menú principal")

        opcion = int(input("Seleccione una opción: "))

        if opcion == len(dulces) + 1:
            print("¡Gracias vuelve pronto!")
            break

        if opcion < 1 or opcion > len(dulces):
            print("Opción inválida, por favor ingrese una opción válida")
            continue

        dulce_seleccionado = dulces[opcion - 1]
        comprar_dulce(dulce_seleccionado)

  elif opcion == 3:
    while True:
        print("¿Qué fritura te gustaría?")
        for i, fritura in enumerate(frituras, start=1):
            print(f"{i}: {fritura.nombre} {fritura.cantidad} ${fritura.precio}")
        print(f"{len(frituras) + 1}: Volver al menú principal")

        opcion = int(input("Seleccione una opción: "))

        if opcion == len(frituras) + 1:
            print("¡Gracias vuelve pronto!")
            break

        if opcion < 1 or opcion > len(frituras):
            print("Opción inválida, por favor ingrese una opción válida")
            continue

        fritura_seleccionada = frituras[opcion - 1]
        comprar_fritura(fritura_seleccionada)
  elif opcion == 4:
    print("¡Gracias vuelve pronto!")
    break

  else:
    print('Error, elija una opción valida')
```
